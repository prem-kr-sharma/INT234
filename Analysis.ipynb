```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
def uni_ordinal(df, col):
    print(f"Univariate Analysis for: {col}")
    print("-" * 40)

    counts = df[col].value_counts().sort_index()
    print(counts)

    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f"Rating Distribution for {col}")
    plt.xlabel("Rating (1 = Very Low, 5 = Very High)")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
```

```python
l = df.select_dtypes(include=['number']).columns

for i in l:
    uni_ordinal(df, i)
    print("\n\n")
```

```python
rest = [x for x in df.columns if x not in l]
rest
```

```python
def uni_categorical(df, col, top_n=10):
    print(f"Univariate Analysis for: {col}")
    print("-" * 40)

    counts = df[col].value_counts(dropna=False)
    print(counts)
    print("\nUnique values:", df[col].nunique())

    # Keep only top N categories
    top_categories = counts.head(top_n)

    plot_df = top_categories.reset_index()
    plot_df.columns = [col, "count"]

    plt.figure(figsize=(8, 4))
    sns.barplot(
        data=plot_df,
        y=col,
        x="count"
    )
    plt.title(f"plot for {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

```python
for i in rest:
  uni_categorical(df, i)
  print('\n\n')
```
